```python
dataset = pd.read_excel('./RHTS_MODE_CHOICE/HH_Public.xlsx')
```

```python
def clean_data(data):
    headings = list(data)
    column_list = []
    for col in headings:
        print("\n")
        print("The current column indexed is ", col)
        res = input("Would you like to drop this column?")
        if res=="y":
            column = str(col)
            column_list.append(column)
            print("I just dropped that for you...", col)
        elif res=="n":
            print("I kept that column for you...", col)
        else:
            print("That was not a valid response... For your safety and sanity I kept that column")
    return column_list
```

```python
column_list = clean_data(dataset)
```

```python
print(column_list)
```

```python
dataset = dataset.drop(column_list, axis=1)
```

```python
len(column_list)
```

```python
dataset.HHVEH.value_counts(dropna=False)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(dataset['HHVEH'], bins = 9)
plt.title('Number of Household Vehicles')
plt.xlabel('Autos')
plt.ylabel('Frequency')
plt.show()
```
